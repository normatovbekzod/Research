```julia
function t(z::Complex)
    if abs(z) <= 1
        return z^(1/3)
    else
        return im*(im*z)^(1/3)
    end
end

ComplexPortraits.phaseplot(-3.0+3.0im, 3.0-3.0im, z -> t(z), point_color=ComplexPortraits.cs_p())
```

```julia
# Define primitive cube root of unity
α=-(1-im*sqrt(3))/2;
```

```julia
# Problem 4: Bekzod Niall Δ
Δ(λ) = -im*λ*(α^2-α) * ( exp(-im*λ) + α*exp(-im*α*λ) + α^2*exp(-im*α^2*λ) )
```

```julia
plotdim=20.0;
plotdispsize=800;
ComplexPortraits.phaseplot(-plotdim+plotdim*im, plotdim-plotdim*im, Δ, point_color=ComplexPortraits.cs_p())
p=plot!(size=(plotdispsize,plotdispsize))
```
